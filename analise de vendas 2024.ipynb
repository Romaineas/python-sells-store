```python
import pandas as pd
```

```python
df= pd.read_excel("Vendas (1).xlsx")
df.head()
```

```python
df.isnull().sum()
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Summarizing data by product
product_sums = df.groupby('Produto')['Quantidade'].sum().reset_index()

# Plotting the bar graph
plt.figure(figsize=(10, 8))
sns.barplot(x='Produto', y='Quantidade', data=product_sums, palette="viridis")

plt.title('Quantidade by Produto')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.tight_layout()
plt.show()
```

```python
pivot_table = df.pivot_table(index='Produto', values='Valor Unit치rio', aggfunc='sum')

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", cbar=True)

plt.title('Heatmap de Quantidade por Valor unit치rio')
plt.xlabel('Valor Unit치rio')
plt.ylabel('Produto')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Produto', y='Quantidade', hue='ID Loja', data=df, palette="viridis")

plt.title('Quantity by Product and Store ID')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.legend(title='ID Loja')
plt.tight_layout()
plt.show()
```
